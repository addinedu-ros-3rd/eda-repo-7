```python
import pandas as pd

drama_df = pd.read_excel("../data/drama_df_in_rank_2023-09-14_ver_7.xlsx")
drama_df.head()
```

```python
sql = "INSERT INTO kdrama_test VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"

for idx in drama_df.index:
    tmp  = list(drama_df.loc[idx].values)
    for i, v in enumerate(tmp):
        tmp[i] = str(v)
    cur.execute(sql, tuple(tmp))
    conn.commit()

cur.execute("SELECT * FROM kdrama_test;")
result = cur.fetchall()
for row in result:
    print(row)

conn.close()
```
